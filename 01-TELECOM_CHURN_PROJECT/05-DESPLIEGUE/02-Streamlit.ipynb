```python
pip install streamlit pandas joblib requests dill
import streamlit as st
import pandas as pd
import joblib
import requests
import dill

# Carga el modelo
# GitHub raw content URL
github_url = 'https://raw.githubusercontent.com/Martinerramuspe/02-DATA_SCIENCIE/main/01-TELECOM_CHURN_PROJECT/05-DESPLIEGUE/Model_prepro.pkl'
response = requests.get(github_url)
Model_prepro = dill.loads(response.content)

# Función para hacer la predicción
def predict(data):
    prediction = Model_prepro.predict(data)
    return prediction[0]

# Configuración de la aplicación
st.title('Aplicación de Predicción')
st.write('Esta aplicación utiliza un modelo de predicción para predecir uno o cero.')

# Definir los datos de entrada
X = {
    'State': 'NY',
    'Account length': 150,
    'Area code': 212,
    'International plan': 'Yes',#vivo.
    'Voice mail plan': 'No',#vivo.
    'Number vmail messages': 0,#vivo.
    'Total day minutes': 300.5,#vivo.
    'Total day calls': 120,
    'Total day charge': 50.75,#vivo.
    'Total eve minutes': 250.3,#vivo.
    'Total eve calls': 80,
    'Total eve charge': 21.15,#vivo.
    'Total night minutes': 200.2,#vivo.
    'Total night calls': 100,
    'Total night charge': 9.01,#vivo.
    'Total intl minutes': 15.5,#vivo.
    'Total intl calls': 4,#vivo
    'Total intl charge': 4.2,
    'Customer service calls': 2,
    'Churn': False
}

# Agregar sliders para los inputs
input_Number_vmail_messages = st.slider('Number vmail messages', min_value=0, max_value=70, value=X['Number vmail messages'])
input_Total_day_minutes = st.slider('Total day minutes', min_value=0, max_value=400, value=X['Total day minutes'])
input_Total_day_charge = st.slider('Total day charge', min_value=0, max_value=80, value=X['Total day charge'])
input_Total_eve_minutes = st.slider('Total eve minutes', min_value=0, max_value=400, value=X['Total eve minutes'])
input_Total_eve_charge = st.slider('Total eve charge', min_value=0, max_value=40, value=X['Total eve charge'])
input_Total_night_minutes = st.slider('Total night minutes', min_value=0, max_value=450, value=X['Total night minutes'])
input_Total_night_charge = st.slider('Total night charge', min_value=0, max_value=20, value=X['Total night charge'])
input_Total_intl_minutes = st.slider('Total intl minutes', min_value=0, max_value=30, value=X['Total intl minutes'])
input_Total_intl_calls = st.slider('Total intl calls', min_value=0, max_value=30, value=X['Total intl calls'])

# Actualizar los valores de entrada
X.update({
    'Number vmail messages': input_Number_vmail_messages,
    'Total day minutes': input_Total_day_minutes,
    'Total day charge': input_Total_day_charge,
    'Total eve minutes': input_Total_eve_minutes,
    'Total eve charge': input_Total_eve_charge,
    'Total night minutes': input_Total_night_minutes,
    'Total night charge': input_Total_night_charge,
    'Total intl minutes': input_Total_intl_minutes,
    'Total intl calls': input_Total_intl_calls
})

# Agregar selección para 'International plan' y 'Voice mail plan'
input_international_plan = st.radio('International plan', ['Yes', 'No'], index=0 if X['International plan'] == 'Yes' else 1)
input_voice_mail_plan = st.radio('Voice mail plan', ['Yes', 'No'], index=0 if X['Voice mail plan'] == 'Yes' else 1)

# Actualizar los valores de 'International plan' y 'Voice mail plan'
X.update({
    'International plan': input_international_plan,
    'Voice mail plan': input_voice_mail_plan
})

# Convertir los datos en un DataFrame
X_df = pd.DataFrame([X])

# Realizar la predicción cuando se presiona el botón
if st.button('Predecir'):
    prediction = predict(X_df)
    st.write('La predicción es:', prediction)
```
