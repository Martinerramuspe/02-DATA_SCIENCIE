# Ejercicio CV para Model Selection:

En el contexto de un problema de clasificacion, usamos el modelo de knn, pero como elegir el k? Usando la crossvalidation decida cual de los siguentes k genera un modelo mas performante.

```python
ks=[3,4,5,6,7,8]
```

Recuerde que la crossvalidation es una tecnica que permite evaluar la performance de un modelo sin tener que usar el set de testeo. Siempre necesitamos tener un set de testeo el cual usaremos al final de la seleccion de las variables o de los hiperparametros (en este caso la 'k').


```python
from sklearn.model_selection import cross_val_score
```

```python
sklearn.model_selection.cross_val_score(modelo, X, y, scoring, cv)
```

**estimator**: modelo

**scoring**: metrica

**cv**: cantidad de folds
