```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Parte 01: Uso de todas las variables.

```python
X= df[['school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu',
       'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'studytime',
       'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
       'higher', 'internet', 'romantic', 'famrel', 'freetime', 'goout', 'Dalc',
       'Walc', 'health', 'absences']]
y= df['G1']
```

```python
import numpy as np
```

```python
X=pd.get_dummies(X,drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
modelo = LinearRegression() # Asignamos nombre
modelo.fit(X_train,y_train) # aprendizaje
y_pred=modelo.predict(X_train) # prediccion
r2_score(y_train, y_pred)
```

```python
# TEST
y_pred_test=modelo.predict(X_test)
r2_score(y_pred_test,y_test)
```

# Parte 02: Uso de 2 variables.
