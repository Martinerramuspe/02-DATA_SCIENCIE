# Metricas algoritmos de clasificacion

````python
 from sklearn.datasets import load_breast_cancer
 from sklearn.model_selection import train_test_split
 from sklearn import metrics
 import pandas as pd
 import numpy as np
 from matplotlib import pyplot as plt
 import seaborn as sns
 sns.set_style('whitegrid')
````

````python
data=sns.load_dataset('titanic')
data=data.drop(['deck','embark_town','alive','adult_male','class','who'],axis=1)
data=data.dropna()
````

````python
# embarked Port of Embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)
# sibsp Number of Siblings/Spouses Aboard
# parch Number of Parents/Children Aboard
````

````python
data
````

````python
data=pd.get_dummies(data,drop_first=True)
````

````python
data=pd.get_dummies(data,drop_first=True)
````

````python
data
````

````python
X=data.drop('survived',axis=1)
y=data.survived
````

````python
X_train,X_test,y_train,y_test = train_test_split(X,y)
````

````python
from sklearn.neighbors import KNeighborsClassifier
````

## Pruebo primer modelo

````python
m1=KNeighborsClassifier(1) # Asignamos nombre
m1.fit(X_train,y_train) # Aprendizaje
ypred1=m1.predict(X_train) # Prediccion
````

````python
from sklearn.metrics import accuracy_score
````

````python
accuracy_score(y_train,ypred1)
````

````python
ypred2=m1.predict(X_test)
accuracy_score(y_test,ypred2)
````

## Pruebo un segundo modelo

````python
# k= 5
m2=KNeighborsClassifier(5)
m2.fit(X_train,y_train)
ypred2=m2.predict(X_train)
````

````python
accuracy_score(y_train,ypred2)
````

````python
ypred3=m2.predict(X_test)
accuracy_score(y_test,ypred3)
````

## Cual elijo? Esta bien esto? Cual es la evaluacion final del modelo elegido?


* El **test** **set** es como los dolares, es lo ultimo que se toca.
* Con el objetivo de no utilizar nuestras test_set para definir la eleccion del modelo. Procedemos a la utilizacion de la herramienta** cross_vaL_score**. Que nos permite particionar el train_set, y hacer pruebas con cada una de esas, logrando asi, medir la eficiencia del modelo.



# Crossvalidation

![](https://cs.calvin.edu/courses/data/202/ka37/slides/w11/w11d1-cv-review_files/figure-html/cv-anim-.gif)


 **`cross_val_score`**



```python
sklearn.model_selection.cross_val_score(estimator, X,
y, scoring=None, cv=None)
```

**estimator**: modelo

**scoring**: metrica

**cv**: cantidad de folds

````python
X_train,X_test,y_train,y_test = train_test_split(X,y)
````

````python
X_train.shape
````

````python
X_test.shape
````

````python
y_test.shape
````

````python
m1=KNeighborsClassifier(1)
````

````python
m2=KNeighborsClassifier(15)
````

````python
from sklearn.model_selection import cross_val_score
````

````python
cross_val_score(m1,X_train,y_train,cv=3) # Particionamos el Train_test en 3.
````

````python
cross_val_score(m1,X_train,y_train,cv=3).mean() #Como nos devuelve un array, podemos aplciar .mean().
````

````python
cross_val_score(m2,X_train,y_train,cv=3).mean()
````

El mejor modelo, es el m2, ya que entrego un 0,68.

````python
m2.fit(X_train,y_train) # Como m2, es le mejor modelo, ahora si lo entrenamos con la totalidad del Train_set
````

````python
y_pred=m2.predict(X_test)
````

````python
accuracy_score(y_pred,y_test)
````
