# OBJETIVO
 Determinar entre los modelos M1 y M2,cual es que entrega resultado mas eficiente, para luego entregarlo a  comitente,y que realice sus pruebas.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# LA FUENTE DE ESTE DATAFRAME ES KAGLE, SE CARGO EL .CSV EN REPOSITORIO DE GITHUB.
url = "https://github.com/Martinerramuspe/02-DATA_SCIENCIE/blob/main/00-CODER_HOUSE/00-DATASETS/student-mat.csv?raw=true"
df = pd.read_csv(url)
df.head()
```

```python
X= df[['school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu',
       'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'studytime',
       'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
       'higher', 'internet', 'romantic', 'famrel', 'freetime', 'goout', 'Dalc',
       'Walc', 'health', 'absences']]
y= df['G1']
```

```python
import numpy as np
X=pd.get_dummies(X,drop_first=True)
```

```python
df['G1_binary'] = (df['G1'] > 13).astype(int)
y=df['G1_binary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
m1=KNeighborsClassifier(3)
```

```python
m2=KNeighborsClassifier(21)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(m1,X_train,y_train,cv=3)
```

```python
cross_val_score(m1,X_train,y_train,cv=3).mean()
```

```python
cross_val_score(m2,X_train,y_train,cv=3)
```

```python
cross_val_score(m2,X_train,y_train,cv=3).mean()
```

# MODELO GANADOR : M2
* Entregamos el modelo al cliente para que realices sus prueba con un set de datos ( Test_X y Test_y).
* El resultado obtenido por el comitente:
  * 0.7676767676767676 ( mejor que lo que obtuvimos nosotros en M2).
  * Cliente contento.

```python
from sklearn.metrics import accuracy_score
```

```python
m2.fit(X_train,y_train)
ypred2=m2.predict(X_test)
```

```python
accuracy_score(y_test,ypred2)
```
