## **QUALIDIADES DEL VINO TINTO**
Se extrajeron mueestras de distintos vinos tintos, para realizarles ensayos fisicoqu√≠micas, logrando asi conseguir datos cuantitativos para cada uno. La calidad del vino se ve definida por estas caracteritiscas mensionadas.
El objetivo de este poder predecir calidad de un vino con las caracteristicas registradas.

```python
import pandas as pd
```

```python
# LA FUENTE DE ESTE DATAFRAME ES KAGLE, SE CARGO EL .CSV EN REPOSITORIO DE GITHUB.
url = "https://raw.githubusercontent.com/Martinerramuspe/02-DATA_SCIENCIE/main/00-CODER_HOUSE/00-DATASETS/winequality-red.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.shape
```

VARIABLES DE ENTRADA (basadas en pruebas fisicoqu√≠micas):
* fixed acidity= acidez fija
* volatile acidity= acidez vol√°til
* citric acid= √°cido c√≠trico
* residual sugar= az√∫car residual
* chlorides= cloruros
* free sulfur dioxide= di√≥xido de azufre libre
* total sulfur dioxide= di√≥xido de azufre total
* density= densidad
* pH
* sulphates= sulfatos
* alcohol
* quality= calidad (puntuaci√≥n entre 0 y 10)

```python
# PODEMOS OBSERVAR QUE EL DATAFRAME CUENTA CON 12 VARIABLES CUANTITATIVAS.
df.info()
```

```python
# DISTRIBUCION DE LOS DATOS NULOS SOBRE LAS VARIABLES.
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(df.isnull(), cbar=False)
plt.show()
# SE PUEDE OBSERVAR QUE CONTAMOS CON DATOS NULOS DENTRO DE LOS DATAFRAME.
```

```python
# CORRELACION QUE EXISTE ENTRE VARIABLES CUANTITATIVAS.
import seaborn as sns
plt.figure(figsize=(10, 8))
matriz = df.corr().round(2)
sns.heatmap(data=matriz, annot=True)
plt.show()
# SE PUEDE VER CORRELACION INDIRECTAS Y DIRECTAS UN POCO FUERTES.
```

```python
# GRAFICAMOS LAS RELACION ENTRE VARIABLES, PARA VER SI DISTRIBUCION CON RESPECTO A SUS EJES, Y SU DISTRIBUCION DE FORMA INDEPENDIENTE
sns.pairplot(df, diag_kind="kde")
plt.show()
```
