## **REGISTRO VENTAS DE VIVIENDAS EN MELBOURNE-AUSTRALIA**
Este dataset cuenta con un conjunto de datos referente a cada venta de propiedades realizadas en la Ciudad de Melbourne-Australia, ya se caracteristicas de la zona,como de la vivienda en si, puediendo asi poder estudiar comportamientos en cada una de las ventas y las relaciones que puede ver entre si.
Es de interes, podes determinar el precio de una vivienda en funcion a sus caracteristicas.

```python
import pandas as pd
```

```python
# LA FUENTE DE ESTE DATAFRAME ES KAGLE, SE CARGO EL .CSV EN REPOSITORIO DE GITHUB.
url = "https://github.com/Martinerramuspe/02-DATA_SCIENCIE/blob/main/00-CODER_HOUSE/00-DATASETS/melb_data.csv?raw=true"
df = pd.read_csv(url)
```

TRADUCCION Y EXPLICACION DE VARIABLES:

* Suburb: Barrio ---------------------------> Nombre del barrio que conforman la ciudad de Melbourne.
* Address: Domicilio --------------------> Ubicacion de la viviendas dentro de la ciudad.
* Rooms: Habitaciones ----------------> N° numero de habitaciones que posee la vivienda.
* Type: Tipo de vivienda ---------------> Clasificacion de la vivienda en: u - unidad-dúplex, t - casa adosada, h - casa-cabaña-villa-semi-terraza.
* Price: Precio ------------------------------> Precio en le cual se vendio la vivieda.
* Method: Metodo de venta----------> S - propiedad vendida; SP - propiedad vendida anteriormente; PI - propiedad transmitida; PN - vendido antes no divulgado; SN - vendido no divulgado; NB: ninguna oferta; VB - oferta del proveedor; W - retirado antes de la subasta; SA - vendido después de la subasta; SS: precio vendido después de la subasta no revelado ; N/A: precio o oferta más alta no disponible.
* SellerG: Agenda -------------------------> Inmobiliaria responsable de la venta.
* Date: Fecha de venta -----------------> Fecha que se ejecuto a venta de la propiedad.
* Distance: Distancia CDB ------------> Distancia a la zona centrica de la ciudad.
* Postcode: Codigo postal------------> Indentificador geografico.
* Bedroom2: Habitaciones -----------> Número extraído de dormitorios (de diferente fuente).
* Bathroom: Baños------------------------> N° de baños.
* Car: Coche ---------------------------------> N° numero de cocheras de disponible.
* Landsize: M2_terreno ----------------> Metros cuadrados del terreno de la vivienda.
* BuildingArea: M2_edificacion ----> Metros cuadrados de la edificacion.
* YearBuilt: Edad_edificacion -------> Año en el cual fue construido la edificacion.
* CouncilArea: Administracion------> Distrito adminstrativo, encargador de parque, residuos,planificacion urbana, etc.
* Lattitude: Latitud -----------------------> Coordenada geografica.
* Longtitude: Longitud------------------> Coordenada geografica.
* Regionname: Area geografica----> Area geografica.
* Propertycount: N°_propiedades-> Número de propiedades que existen en el suburbio.

```python
# CORRELACION QUE EXISTE ENTRE VARIABLES CUANTITATIVAS.
import seaborn as sns
plt.figure(figsize=(10, 8))
matriz = df.corr().round(2)
sns.heatmap(data=matriz, annot=True)
plt.show()
# OBSERVACION: SE PUEDE QUE LOS VALORES SON MUY DEBILES.
```

```python
# GRAFICAMOS LAS RELACION ENTRE VARIABLES, PARA VER SI DISTRIBUCION CON RESPECTO A SUS EJES, Y SU DISTRIBUCION DE FORMA INDEPENDIENTE
sns.pairplot(df, diag_kind="kde")
plt.show()
```
